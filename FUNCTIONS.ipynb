  -ARGUMENTS:Arguments are the actual values or data you provide to a function when calling it.these values are assigned to the corresponding parameters.
  -python supports several types of arguments:

  1-POSITIONAL ARGUMENTS:Passed in order and assigned to parameters by their position.

  2-DEFAULT ARGUMENT:parameters with default values.if no argument is provided ,the default is used.

  3-VARIABLE LENGTH ARGUMENT:when you don't know the no. of agruments.
  
  4-KEYWORD ARGUMENT:passed using the parameter name,allowing you to specify values out of order.

```python
#POSITIONAL ARGUMENT
def addition(a,b):
  return a+b
addition(5,6)
```

```python
#DEFAULT ARGUMENT
def sum1(b,c,a=0):
  return a+b+c
sum1(b=3,c=5)

```

```python
#VARIALE LENTH ARGUMENT
def sum1(*args):
  s=0
  for i in args:
    s=s+i
  return s
sum1(1,2,3,4,5,6,7)
```

```python
#KEYWORD ARGUMENT
def marks_in_subject(**kwargs):
  marks_list=[]
  for sub,marks in kwargs.items():
    marks_list.append(marks)
  return marks_list
marks_in_subject(a=90,b=30)
```

6. Explain the concept of generators in Python and how they are defined.

  -In python generators are a type of iterable ,like lists or tuples,that allow you to produce a sequence of values lazliy(one at a time)instead of computing them all at once and storing them in memory.this makes them more memory

  -efficient,especially when working with large datasets or infiite sequnces.A generator function is a regular function that uses the yield statement to produce values one at a time.
  
  -The yield statement pauses the funcion,saving its state.when next()is called on the generator ,execution resumes from where it left off.the function stops when there are no more yield statemnets or when it explicity ends with return.


```python
#REGULAR FUNCTION
def square_number(n):
  result=[]
  for i in n:
    result.append(i**2)
  return result
square_number([5,10,15])
```

```python
#YIELD FUNCTION
def square_number_generators(n):
  for i in range(n) :
    yield i**2
square_number_generators(10)
```

```python
gen=square_number_generators (10)
gen
```

```python
next(gen)
```

```python
next(gen)
```

7. What are the advantages of using generators over regular functions?

  -Using generators instead of regular functions can provide several advantages.

  1.generators produce values on demand rather than computing all values upfront .this can save memory and computation time when working with large data sets or infinite sequences.

  2.generators use an iterator protocol,yielding one value at a time and not storing the entire sequence in memory.

  3.generators avoid the overhead of creating and storing large lists or data structures in memory.

  4.regular functions can't easily represent infinite sequences as they would require infinite memory.generators,however can produce infinite values one at a time without issues.
  
  5.generators can be chained together to create pipelines for processing data.

```python
def square_number(n):
  result=[]
  for i in n:
    result.append(i**2)
  return result
square_number([5,10,15])
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
  """
  returns the sum of all even numbers in the list.
  :param numbers:list of integers
  :return:sum of even numbers
  """
  return sum(num for num in numbers if num % 2 == 0)
example_list =[1,2,3,4,5,6]
print(sum_of_even_numbers(example_list))
```

2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse_string(input_string):
  """
  accepts a string and returns its reverse.
  parameters:
  input_string(str):the string to be reversed.
  returns:
  str: the reversed string.
  """
  return input_string[::-1]
text="hello world!"
reversed_text = reverse_string(text)
print("reversed string:",reversed_text)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32
celsius_temperature = [0,20,30,40,100]
fahrenheit_temperature = list(map(celsius_to_fahrenheit,celsius_temperature))
print("celsius temperature:",celsius_temperature)
print("fahrenheit temperature:",fahrenheit_temperature)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
  def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels
  result = filter(is_not_vowel,input_string)
  return "".join(result)
input_string = input("enter a string:")
output_string = remove_vowels(input_string)
print("string without vowels:",output_string)
```
